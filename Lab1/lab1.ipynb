# lab-1: Problem solving using searching

```python
def is_safe(board, x, y, n):
    #column test
    for row in range(x):
        if board[row][y] == 'Q':
            return False



    #Left upper diagonal test
    r = x - 1
    c = y - 1
    while(r >= 0 and c >= 0):

        if(board[r][c] == 'Q'):
            return False
        r = r - 1
        c = c - 1
    
            


    #Right upper diagonal test
    r = x - 1
    c = y + 1
    while(r >= 0 and c < n):

        if(board[r][c] == 'Q'):
            return False
        
        r = r - 1
        c = c + 1

    return True

```

```python
from pprint import pprint
n = 4
board = [['_']*n for i in range(n)]
board[0][0] = 'Q'
pprint(board)
print(is_safe(board,1,0,n))
```

```python
n = int(input("Enter the number of queens: "))
board = [['_']*n for i in range(n)]
def nQueenSolver(board, row, n):
    if (row>=n):
        return True
    
    for col in range(n):
        if is_safe(board, row, col, n):
            board[row][col] = 'Q'
            if(nQueenSolver(board, row+1, n)):
                return True
            board[row][col] = '_'

    return False

if(nQueenSolver(board,0,n)):
    pprint(board)

else:
    pprint("No solution exists....")
```
