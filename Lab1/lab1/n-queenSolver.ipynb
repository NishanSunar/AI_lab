### Lab1: Problem solving by search
#### 1. N-Queens problem

```python
def is_safe(board, x, y, n):
    # Column test
    for row in range(x):
        if board[row][y] == 'Q':
            return False
        
    # left upper diagonal test
    r = x - 1
    c = y - 1
    while(r >= 0 and c >= 0):
        if board[r][c] == 'Q':
            return False
        r -= 1
        c -= 1

    # right lower diagonal test
    r = x - 1
    c = y + 1
    while(r >= 0 and c < n):
        if board[r][c] == 'Q':
            return False
        r -= 1
        c += 1

    # return true if all the tests are passed
    return True
```

```python
from pprint import pprint

n = 4
board = [['_'] * n for i in range(n)]
board[3][0] = 'Q'
pprint(is_safe(board = board, x = 0, y = 3, n = n))
board
```

```python
n = int(input("Enter number of Q's: "))
board = [['_'] * n for i in range(n)]
def nQueensSolver(board, row, n):
    if row >= n:
        return True

    for col in range(n):
        if is_safe(board, row, col, n):
            board[row][col] = 'Q'
            if nQueensSolver(board, row + 1, n):
                return True
            board[row][col] = '_'
    return False

if(nQueensSolver(board, 0, n)):
    pprint(board)
else:
    print("No solution exists")
```

#### Time complexity: O(n!)
#### Space complexity: O(n^2)
